```python
# drop path column
Chexpert_train_df.drop(columns=['Sex'], inplace=True)
Chexpert_test_df.drop(columns=["Sex"], inplace=True)
Chexpert_validation_df.drop(columns=["Sex"], inplace=True)

# replace column names
column_mapping = {
    'patient_id': 'subject id',
    'age_decile': 'age decile',  # already matches, included for clarity
    'GENDER': 'gender',  # Duplicate, to be resolved
    'Race': 'race',
}
Chexpert_train_df.rename(columns=column_mapping, inplace=True)
Chexpert_test_df.rename(columns=column_mapping, inplace=True)
Chexpert_validation_df.rename(columns=column_mapping, inplace=True)
```

```python
# in gender column make Male M and Female F
Chexpert_train_df["gender"] = Chexpert_train_df["gender"].apply(lambda x: 'M' if x == "Male" else 'F')
Chexpert_test_df["gender"] = Chexpert_test_df["gender"].apply(lambda x: 'M' if x == "Male" else 'F')
Chexpert_validation_df["gender"] = Chexpert_validation_df["gender"].apply(lambda x: 'M' if x == "Male" else 'F')
```

```python

column_mapping = {
    'White': 'WHITE',
    'Other': 'OTHER',
    'Asian': 'ASIAN',
    'Black': 'BLACK/AFRICAN AMERICAN',
    'Hispanic': 'HISPANIC/LATINO',
    'American Indian/Alaska native': 'AMERICAN INDIAN/ALASKA NATIVE',
}
Chexpert_train_df["race"] = Chexpert_train_df["race"].apply(lambda x: column_mapping[x] if x in column_mapping else x)
Chexpert_test_df["race"] = Chexpert_test_df["race"].apply(lambda x: column_mapping[x] if x in column_mapping else x)
Chexpert_validation_df["race"] = Chexpert_validation_df["race"].apply(lambda x: column_mapping[x] if x in column_mapping else x)

```

```python
minic_train_df["path"] = minic_train_df["path"].apply(lambda x: x.replace(".tfrecord", ""))
minic_test_df["path"] = minic_test_df["path"].apply(lambda x: x.replace(".tfrecord", ""))
mimic_validation_df["path"] = mimic_validation_df["path"].apply(lambda x: x.replace(".tfrecord", ""))

Chexpert_train_df["path"] = Chexpert_train_df["path"].apply(lambda x: x.replace(".jpg", ""))
Chexpert_test_df["path"] = Chexpert_test_df["path"].apply(lambda x: x.replace(".jpg", ""))
Chexpert_validation_df["path"] = Chexpert_validation_df["path"].apply(lambda x: x.replace(".jpg", ""))
```

```python

# List of yaml files
yaml_files = [
  "image_all_disease_classification_19.yaml",
  "image_all_disease_classification_31.yaml",
  "image_all_disease_classification_38.yaml",
  "image_all_disease_classification_47.yaml",
  "image_all_disease_classification_77.yaml",
  "image_all_race_detection.yaml",
  "image_all_gender_detection.yaml",
  "vector_all_disease_classification_19.yaml",
  "vector_all_disease_classification_31.yaml",
  "vector_all_disease_classification_38.yaml",
  "vector_all_disease_classification_47.yaml",
  "vector_all_disease_classification_77.yaml",
  "vector_all_gender_detection.yaml",
  "vector_all_race_detection.yaml",
  "vector_chexpert_disease_classification_19.yaml",
  "vector_chexpert_disease_classification_31.yaml",
  "vector_chexpert_disease_classification_38.yaml",
  "vector_chexpert_disease_classification_47.yaml",
  "vector_chexpert_disease_classification_77.yaml",
  "vector_chexpert_gender_detection.yaml",
  "vector_chexpert_race_detection.yaml",
  "vector_mimic_disease_classification_19.yaml",
  "vector_mimic_disease_classification_31.yaml",
  "vector_mimic_disease_classification_38.yaml",
  "vector_mimic_disease_classification_47.yaml",
  "vector_mimic_disease_classification_77.yaml",
  "vector_mimic_gender_detection.yaml",
  "vector_mimic_race_detection.yaml"
]

import os
import subprocess

import os
import re

def find_least_loss_ckpt_file(model_dir):
    files = os.listdir(model_dir)
    ckpt_files = [file for file in files if file.endswith('.ckpt')]
    # Adjust the regular expression to match the provided sample format
    loss_pattern = re.compile(r'val_loss=([0-9]+\.[0-9]+).ckpt')

    file_losses = {}

    for file in ckpt_files:
        match = loss_pattern.search(file)
        if match:
            file_losses[file] = float(match.group(1))

    if file_losses:
        return min(file_losses, key=file_losses.get)
    else:
        return None

# Use this adjusted function in `create_bash_script` in place of `find_latest_ckpt_file`


def create_bash_script(yaml_file):
    # Determine the model directory based on yaml_file name
    model_dir = f"/fs01/home/hhamidi/diff/fariness_embedding_second_submit/fairness_on_embeddings/results/{yaml_file.replace('.yaml', '')}/checkpoints"
    # Find the latest checkpoint file in the model directory
    ckpt_file_name = find_least_loss_ckpt_file(model_dir)
    print("python model_evaluation.py {model_dir}/".format(model_dir=model_dir))

    
    if ckpt_file_name:
        script_content = f'''#!/bin/bash

#SBATCH --gres=gpu:rtx6000:1
#SBATCH --qos=m5
#SBATCH --time=1:00:00
#SBATCH -c 39
#SBATCH --mem=90G
#SBATCH --job-name={yaml_file.replace('.yaml', '')}
#SBATCH --output=./results/{yaml_file.replace('.yaml', '.log')}

echo Running on $(hostname)

# activate environment
source /h/hhamidi/.bashrc
conda activate diff
cd /fs01/home/hhamidi/diff/fariness_embedding_second_submit/fairness_on_embeddings

# put your command here
python main.py predict -c ./configs/vector/{yaml_file} --ckpt_path {model_dir}/{ckpt_file_name}
'''
        with open(f"./sbatch/test/{yaml_file.replace('.yaml', '_run.sh')}", 'w') as f:
            f.write(script_content)
    else:
        print(f"No checkpoint file found for {yaml_file.replace('.yaml', '')}")

# List of yaml files

# Create bash script for each yaml file
for yaml_file in yaml_files:
    create_bash_script(yaml_file)

```
